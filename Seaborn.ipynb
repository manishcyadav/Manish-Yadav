# Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd



var = [1,2,3,4,5,6,7]
var_1=[2,3,4,1,5,6,7]

x_1 = pd.DataFrame({"var":var,"var_1":var_1})

sns.lineplot(x="var",y="var_1",data=x_1)

plt.show()
```

```python
# How To Load dataset from github to jupyter

y_1=sns.load_dataset("penguins").head(20)
y_1
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=y_1,hue="sex",size=15,
             style="sex",palette='Accent',markers=["o",">"],dashes=True,legend=True)


# Hue parameter determine which column in DataFrame should be used for color encoding
# Style parameter is used to change the size of the particular column
# Palette parameter is used to change the color of the plot
# marker parameter is used to show the markers on the plot
# Dashes parameter is used to remove the dashes on plot

# In Matplotlib, the legend function is used for assigning the labels But in seaborn it is used to remove the labels
plt.grid()
plt.title("Python")
plt.show()
```

# Bar Plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var = sns.load_dataset("penguins")
var
```

```python
sns.barplot(x=var.island,y=var.bill_length_mm)
plt.show()
```

```python

sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex")

# Here,Hue parameter is used to seprate the bars based on the sex
plt.show()
```

```python
order_1  = ["Dream","Torgersen","Biscoe"]
    
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",order=order_1,
           hue_order=["Female","Male"],ci=80)
# ci parameter is used to adjust the lenght of small lines/bars in blach color present above bars ( range is 0 to 100 )
plt.show()
```

```python
# To make horizontal graph, both the parameters must be of same type ex. neumerical

order_1  = ["Dream","Torgersen","Biscoe"]
    
sns.barplot(x="bill_depth_mm",y="flipper_length_mm",data=var,orient="h")

plt.show()
```

```python
sns.set(style="darkgrid")

order_1  = ["Dream","Torgersen","Biscoe"]
    
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",order=order_1,color="m",
           palette="icefire",saturation=0.3,errcolor="b",capsize=0.2)
# Palette is used to give diff colors to the plot
# Saturation is used to adjust the color of the plot
plt.show()
```

# HeatMap

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
var = np.linspace(1,10,20).reshape(4,5)
var
```

```python
y = {"fontsize":20,"color":"r"}

v = sns.heatmap(var,vmin=0,vmax=10,cmap="PuOr",annot=True,annot_kws=y,linewidth = 5, linecolor="black",cbar=True
           ,xticklabels=True,yticklabels=True)
# Vmin and Vmax is used to change the range of the heatmap
# cmap is used to adjust the color of the plot
# To Show the Values on each box, annot parameter is used . 
# annot_kws is used to assign the steps oy y 
# Linecolor and linewidth is used for performing operations on the boundry of plot
# Cbr parameter is used to add or remove the color bar
# xticklabel and yticklabel is used to add or remove the x axis anf y axis values
v.set(xlabel="python",ylabel="DATA")
plt.show()
```

# Box Plot

# Styling Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
sns.set_style("white")

plt.figure(figsize=(12,8))

sns.barplot(x="day",y="total_bill",data=var)
sns.despine()
# despine parameter is used to remove the axis lines
plt.show()
```
