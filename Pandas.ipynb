```python
# USe of DROPNA()

df = pd.DataFrame({"name" : ['manish','atharv','Sagar'],
                    "TOY" : [np.nan ,'Mobile','Buds'],
                   "Born" : [pd.NaT , pd.Timestamp("1940-04-25"),pd.NaT],
                  })
                        
```

```python
df
```

```python
df.dropna()
```

```python
df.dropna(how='all')  # drop row if all na are present

# To remove the colunms we have to give axis ex. df.dropna(how = 'All',axis = 1)
```

```python
# To Remove Duplicates

df = pd.DataFrame({"name" : ['manish','atharv','manish'],
                    "TOY" : [np.nan ,'Mobile','Buds'],
                   "Born" : [pd.NaT , pd.Timestamp("1940-04-25"),pd.NaT],
                  })
                        
```

```python
df
```

```python
df.drop_duplicates(subset="name")
```

```python
df.drop_duplicates(subset="name",keep="first")
```

```python
df.drop_duplicates(subset="name",keep="last")
```

```python
df.drop_duplicates(subset="name",keep=False)
```
