# Bar Plot

```python
x=["Python","C","C++","Java"]
y=[85,70,60,80]
# Modification on bar graph
c=["r","c","y","g"]
plt.xlabel("Languages",fontsize=15)
plt.ylabel("numbers",fontsize=15)
plt.title("Programming",fontsize=20)
plt.bar(x,y,color=c,width=0.4,align="edge",edgecolor="black",linewidth=3,alpha=1,label="popularity")
plt.legend()
plt.show()

# width is used for size of the bars
# align is used for where to place the x axis label

# Linewidth is used for size of the edge with back or any ither color 
# alpha is used to make graph faint to dark(0 to 1)

#label is assigned using the plt.legend().
```

```python
# To draw multiple graphs in a single chart

x=["Python","C","C++","Java"]
y=[85,70,60,80]
z=[20,30,40,50]

c=["r"]
plt.xlabel("Languages",fontsize=15)
plt.ylabel("numbers",fontsize=15)
plt.title("Programming",fontsize=20)
plt.bar(x,y,color=c,width=0.4,align="edge",edgecolor="black",linewidth=3,alpha=1,label="popularity")
plt.bar(x,z,color=["m"],width=0.4,align="edge",edgecolor="black",linewidth=3,alpha=1,label="popularity1")
plt.legend()
plt.show()
```

```python
# To place the 2 graphs at the adjacent side
import matplotlib.pyplot as plt
import numpy as np

x=["Python","C","C++","Java"]
y=[85,70,60,80]
z=[20,30,40,50]

width=0.2
p=np.arange(len(x))            # this is the importtant part
p1=[j+width for j in p]


plt.xlabel("Languages",fontsize=15)
plt.ylabel("numbers",fontsize=15)
plt.title("Programming",fontsize=20)

plt.bar(p,y,width,color=["r"],label="popularity")
plt.bar(p1,z,width,color=["m"],label="popularity1")

# by using this there are numbers present at the base of grapg , if we want to make it labels then do following

plt.xticks(p+width/2,x,rotation=10)   # This is important part

plt.legend()
plt.show()
```

```python
# To make horizontal Bar Graph use barh

import matplotlib.pyplot as plt
import numpy as np

x=["Python","C","C++","Java"]
y=[85,70,60,80]
z=[20,30,40,50]

width=0.2
p=np.arange(len(x))            # this is the importtant part
p1=[j+width for j in p]


plt.xlabel("Languages",fontsize=15)
plt.ylabel("numbers",fontsize=15)
plt.title("Programming",fontsize=20)

plt.barh(x,y,width,color=["r"],label="popularity")
plt.barh(x,z,width,color=["m"],label="popularity1")

# by using this there are numbers present at the base of grapg , if we want to make it labels then do following

plt.xticks(p+width/2,x,rotation=10)

plt.legend()
plt.show()
```

# Scatter plot

```python
import matplotlib.pyplot as plt
```

```python
Day=[1,2,3,4,5,6,7]
No=[2,3,4,1,6,7,8]

plt.scatter(Day,No,color="r")

plt.title("Scatter plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("No",fontsize=20)

plt.show()
```

```python
Day=[1,2,3,4,5,6,7]
No=[2,3,4,1,6,7,8]

c=["r","y","b","r","r","m","b"]
s=[100,182,148,179,192,149,169]

plt.scatter(Day,No,color=c,sizes=s,alpha=0.7,marker="*",edgecolor="black",linewidth=2)

plt.title("Scatter plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("No",fontsize=20)

plt.show()
```

```python
# To give color to the scatters by using numbers,folllowing method can be used
import matplotlib.pyplot as plt

Day=[1,2,3,4,5,6,7]
No=[2,3,4,1,6,7,8]

c=[13,56,78,35,60,90,38]  #lies between 0 to 100
s=[100,182,148,179,192,149,169]

plt.scatter(Day,No,c=c,s=s,cmap="BrBG")
plt.colorbar(label="color scale")

plt.title("Scatter plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("No",fontsize=20)

plt.show()
```

```python
# Combination of scatter plot can be done in following way

import matplotlib.pyplot as plt

Day=[1,2,3,4,5,6,7]
No=[2,3,4,1,6,7,8]
No2=[3,2,4,5,6,2,4]
c=[13,56,78,35,60,90,38]  #lies between 0 to 100
s=[100,182,148,179,192,149,169]

plt.scatter(Day,No,c=c,s=s,cmap="BrBG")
plt.scatter(Day,No2,color="r",s=s,cmap="BrBG")
plt.colorbar(label="color scale")

plt.title("Scatter plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("No",fontsize=20)

plt.show()
```

# Pie Plot

```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40]
y=["c","c++","java","Python"]
ex=[0.4,0.0,0.0,0.0]
c=["r","g","y","m"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.2f%%",shadow=True,radius=1.2)   
# explode parameter is used to show the particular part of the pie diffrently
                                                      
# autopct is used to show the percentage of the pie
                                                        
# 0.1 tells that there is 1 no after point
                                                      
# shadow parameter is used to show shadow of the pie
plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","java","Python"]
ex=[0.4,0.0,0.0,0.0]
c=["r","g","y","m"]
plt.pie(x,labels=y,explode=ex,colors=c,shadow=True,autopct="%0.2f%%",labeldistance=1.2,startangle=90,
        textprops={"fontsize":10},counterclock=False,wedgeprops={"linewidth":3,"edgecolor":"black"},rotatelabels=False)  
# The labeldistance assign the label according to the distance
# The startangle tells the starting point of the pie
# textprops assign the size of the labels
# counterclock gices the direction of elements in pie 
# Wedgeprops is used to adjust the line edge
# rotate labels is used to roatete the labels
plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","java","Python"]
ex=[0.4,0.0,0.0,0.0]
c=["r","g","y","m"]
plt.pie(x,labels=y,explode=ex,colors=c,shadow=True,autopct="%0.2f%%",labeldistance=1.2,startangle=90,radius=0.8,
        textprops={"fontsize":7},counterclock=False,wedgeprops={"linewidth":3,"edgecolor":"black"},rotatelabels=False)  
# The labeldistance assign the label according to the distance
# The startangle tells the starting point of the pie
# textprops assign the size of the labels
# counterclock gices the direction of elements in pie 
# Wedgeprops is used to adjust the line edge
# rotate labels is used to roatete the labels
plt.title("Languages")
plt.legend(loc=2)
plt.show()
```

# Box plot & Whisker Plot

```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40,50,60,70,120]

plt.boxplot(x,notch=False,vert=True,widths=0.2,tick_labels=["Python"],patch_artist=False,
           showmeans=True,whis=1,sym="r+",boxprops=dict(color="m"),capprops=dict(color="r"),
           whiskerprops=dict(color="r")) 
# Notch is used to change the shape of the box
# vert= true gives vertical box plot
# Widths parameter is used to change the width of the box
# Patch_artist is the parameter used for assigning the color to the box
# showmeans is used to show the middle part of the box

# The Outler can be made by incresing value of list bye some large amount, in this example we have incr it from 70 to 120

# If we want to join the bar above box to the outler, the make use of whis parameter

# TO change the color or shape of the outler,sym parameter is Used.

# to change the features of Box , boxprops parameter is used.
# to change the features of cap(lines present above box), capprops parameter is used.       # Horizontal lines are known as cap and Vertical as whisk
plt.show()  
```

```python
# To plot Multiple box plots

x=[10,20,30,40,50,60,70]
x1=[10,20,40,50,30,60,90]
y=[x,x1]
plt.boxplot(y,notch=False,vert=True,widths=0.2,tick_labels=["Python","c++"],patch_artist=False,
           showmeans=True,whis=1,sym="r+",boxprops=dict(color="m"),capprops=dict(color="r"),
           whiskerprops=dict(color="r")) 

plt.show()
```

# Stack Plot(Area Plot)

```python
x=[1,2,3,4,5]
area=[1,6,3,4,5]

plt.stackplot(x,area)

plt.show()
```

```python
# Multiple Stack Plot

x=[1,2,3,4,5]
area1=[1,6,3,4,5]
area2=[1,4,2,7,3]
area3=[1,2,3,3,4]

l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=["m","g","y"])

plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(loc=0)
plt.show()
```

# Fill-Between Plot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.plot(x,y,color="r")

plt.fill_between(x=[2,4],y1=2,y2=4,color="y")


plt.title("Python")


plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.show()
```

```python
# Use of Where function
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])

plt.plot(x,y,color="r")

plt.fill_between(x,y,color="g",where=(x>=2)&(x<=4),alpha=0.5)   # Alpha parameter is used to adjust the brightness of the plot

plt.title("Python")


plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.show()
```

# Savefig

```python
# Savefig is the function used to save the images of the plot in folder

y=[1,2,3,4]
plt.subplot(2,2,1)
plt.plot(x,y,color="r")

plt.subplot(2,2,2)
plt.pie([1],colors="r")

x1=[10,20,30,40]
plt.subplot(2,2,3)
plt.pie(x1)

                             # here 1,2,3,4 are the positions of the plots
x2=["a","s","d","t"]
y2=[10,20,30,40]
plt.subplot(2,2,4)
plt.bar(x2,y2)

plt.savefig("Subplot")
plt.show()
```

```python
# by using DPI that is Dot Per  Inch, we can improve the quality of the graph
# By using Facecolor function, we can change the background color pf the fig
# To save the image in the form of PDF then .pdf,.jpg,.png is used(by default get png formate)
# Transparent function can be used to make graph transperant in the sense of True False
# We can change the inches of box using bbox_inches function
y=[1,2,3,4]
plt.subplot(2,2,1)
plt.plot(x,y,color="r")

plt.subplot(2,2,2)
plt.pie([1],colors="r")

x1=[10,20,30,40]
plt.subplot(2,2,3)
plt.pie(x1)

                             # here 1,2,3,4 are the positions of the plots
x2=["a","s","d","t"]
y2=[10,20,30,40]
plt.subplot(2,2,4)
plt.bar(x2,y2)

plt.savefig("Subplot4",dpi=2000,facecolor="y",transparent=True,bbox_inches="tight")      # Change the fname according o requirment
plt.show()
```
